```python
import torch
import torch.onnx
from torch.autograd import Variable
```

```python
from modeling.deeplab import DeepLab
```

```python
# A model class instance
model = DeepLab('mobilenet')
model = model.cuda()
```

```python
# Load the weights from a file (.pth usually)
weights_pth = "checkpoint.pth.tar"
checkpoint = torch.load(weights_pth)
# checkpoint['state_dict']
```

```python
model.load_state_dict(checkpoint['state_dict'])
```

```python
dummy_input = Variable(torch.randn(1, 3, 1080, 1920)).cuda()

torch.onnx.export(model, dummy_input, "m2.onnx", export_params=True, verbose=True)
```
